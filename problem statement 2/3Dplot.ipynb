```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import transforms
```

# A2C in 3D

```python


x = []
y = []
z = []

with open("data/IM_0013-01_inx.txt", "r") as file:
    for line in file:
        point = line.split()
        y.append(float(line.split()[0]))
        z.append(float(line.split()[1]))
        x.append(63.5)
X = np.array(x, dtype=float)
Y = np.array(y, dtype=float)
Z = np.array(z, dtype=float)

    
```

```python
ax = plt.axes(projection='3d')
ax.scatter3D(X, Y, Z)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('A2C in 3d')
```
