# 3-D Plots

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
plt.axes(projection='3d')
```

```python
x = []
y = []
z = []
plt.style.use('seaborn')
with open("data/IM_0013-01_inx.txt", "r") as file:
    for line in file:
        point = line.split()
        y.append(float(line.split()[0]))
        z.append(float(line.split()[1]))
x.append(0.0)
```

```python
ax = plt.axes(projection='3d')
ax.scatter3D(x, y, z)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```

```python
ax = plt.axes(projection='3d')
ax.scatter3D(x, y, z)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(0, 0)
```
